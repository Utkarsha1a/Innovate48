```python
import re
from docx import Document
from dateutil import parser
def extract_word_after_maturity_date(docx_file):
    # Load the Word document
    doc = Document(docx_file)

    # Extract all text from the document
    text = '\n'.join([paragraph.text for paragraph in doc.paragraphs])

    # Search for the word after the "Maturity Date" line using a regular expression pattern
    pattern = r"Maturity\s+Date\*?:\s+([\w\s,]+)"
    match = re.search(pattern, text)


    # If a match is found, extract the word
    if match:
        line = match.group(1).strip()
        words = line.split()

        # Iterate through words in the line
        for word in words:
            # Try to parse the word as a date
            try:
                date = parser.parse(word, fuzzy=True).date()
                return date.strftime("%B %d, %Y")
            except ValueError:
                pass
        return word

    # Return None if the "Maturity Date" line or the subsequent word is not found
    return None

# Specify the Word document file path
docx_file = 'output.docx'

# Extract the word after the Maturity Date
word = extract_word_after_maturity_date(docx_file)

# Print the extracted word
if word:
    print(word)
else:
    print("Maturity Date or the subsequent word not found in the document.")
```
